# Hadoop vs Spark for Carrier Delay

```python
import numpy as np
import matplotlib.pyplot as plt


# Defining number of iteration and Data
x = ["1", "2", "3", "4", "5"]
hadoop_data = [13.433,7.813,7.695,7.916,7.15]

spark_data = [2.3975,0.5118,0.4912,0.7235,0.5473]

barWidth = 0.35

# Set position of bars on X axis
r1 = np.arange(len(hadoop_data))
r2 = [x + barWidth for x in r1]

# Create double bar graph
plt.bar(r1, hadoop_data, color='#0F52BA', width=barWidth, label='Hadoop')
plt.bar(r2, spark_data, color='#73C2FB', width=barWidth, label='Spark')

#  Add labels, title, and legend
plt.xlabel("Number of iterations")
plt.ylabel("Running Time (s)")
plt.title("Hadoop vs Spark for Carrier Delay")
plt.legend()

# Show graph
plt.show()
```
